```python
from bs4 import BeautifulSoup
import requests
import openpyxl
import datetime
wb=openpyxl.Workbook()
ws =wb.active
job ="大數據"
url = "https://m.1111.com.tw/search/job?ks="+job
html = requests.get(url)

ws['A1']="職缺名稱"
ws['B1']="職缺公司"
ws['C1']="職缺連結"
ws['D1']="職缺地址"
ws['E1']="職缺內容"
ws['F1']="職缺經驗"
ws['G1']="職缺學歷"
ws['H1']="計算薪水"
ws['I1']="薪資上限"
ws['J1']="薪資下限"

hrefs = 1
a=0
while a != 1:
    soup =html.text
    soup = BeautifulSoup(soup,'html.parser')
    aaa = soup.find("div",class_="srh-footer__content-item srh-footer__content-nextpage")
    if aaa == None:
        a=1
        break;
    else:
        item = soup.find_all("div",class_ ="item__job")
    
    for a in item:
        title = a.find("a",class_="item__job-info--link item__job-position0--link")
        itemx = a.find_all("i")
        com = a.find("div",class_="item__job-organ").a["title"]
        content =a.find("div",class_="item__job-desc item__job-desc-un_extension")
        txt = title.text.replace("\x08","").replace("\x1a","")
        計算薪水=itemx[1]["aria-label"][:2]
        薪資上限=""
        薪資下限=""
        x=0
        for a in itemx[1]["aria-label"]:
#             print(a)
            
            if a=="0" or a=="1" or a=="2" or a=="3" or a=="4" or a=="5" or a=="6" or a=="7" or a=="8" or a=="9" or a=="萬" or a=="." or a=="~":
                if x==1:
                    薪資上限= 薪資上限+a
                elif a=="~":
                    x=1
                elif x==0:
                    薪資下限 =薪資下限+a 
        if 薪資上限 == "":          
            薪資上限 = 薪資上限+薪資下限
        
        elif "萬" in 薪資上限:
            薪資上限=薪資上限.replace("萬","")
            薪資上限=int(float(薪資上限)*10000)
            

        elif "萬" in 薪資下限:
            薪資下限=薪資下限.replace("萬","")
            薪資下限=int(float(薪資下限)*10000)
        else:
            薪資上限 = int(薪資上限)
            薪資下限 = int(薪資下限)
        print(itemx[1]["aria-label"])
        print("薪資上限",薪資上限)
        print("薪資下限",薪資下限)
        end = [
            txt,
            com,
            title["href"],
            itemx[0]["aria-label"],
            content["title"],
            itemx[2]["aria-label"],
            itemx[3]["aria-label"],
            計算薪水,
            薪資上限,
            薪資下限
        ]  
#         print(end)
        ws.append(end)
    hrefs +=1
    href1 = url+"&page=%s"%(str(hrefs))
    html = requests.get(href1)
    print(href1)

wb.save("1111人力銀行職缺"+str(datetime.date.today())+".xlsx")
print("finish")
```
